```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('Speed_Dating_Data.csv')
```

```python
data.dropna()
```

```python
# df = DataFrame(data, columns = ['iid','pid','gender','match','dec_o',\
#                                 'age','age_o','samerace','int_corr','date','go_out', 'field_cd',\
#                                 'attr3_1','sinc3_1','intel3_1','fun3_1','amb3_1',\
#                                 'attr_o','sinc_o','intel_o','fun_o','amb_o','shar_o',\
#                                 'attr1_1','sinc1_1','intel1_1','fun1_1','amb1_1','shar1_1'])
#only include subjective and objective profile
df = DataFrame(data, columns = ['iid','pid','gender','match','dec_o',\
                                'age','age_o','samerace','int_corr','date','go_out', 'field_cd',\
                                'attr3_1','sinc3_1','intel3_1','fun3_1','amb3_1',\
                                'attr_o','sinc_o','intel_o','fun_o','amb_o','shar_o'])
```

```python
df_m = df[df['gender']==1]#male
df_f = df[df['gender']==0]#female
```

```python
df_mean_m = df_m.groupby('iid').mean()
df_mean_f = df_f.groupby('iid').mean()
```

```python
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
f, ax = plt.subplots(figsize=(11, 9))
cmap = sns.diverging_palette(220, 10, as_cmap=True)
sns.heatmap(corr, mask=mask,cmap=cmap, square=True, linewidths=.5, cbar_kws={"shrink": .5})
f.savefig("corr.png")
```
